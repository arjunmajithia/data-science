```python
import bs4
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
response = requests.get(url)
print(response.content)
```

```python
# parsing the html
soup = bs4.BeautifulSoup(response.content)
```

```python
article_element = soup.findAll('div', {'class': 'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'})
```

```python
article_element[0]
```

```python
article = article_element[0]
```

```python
article.find('img')                      # finds single occurence of img tag inside the div
```

```python
article.img.attrs
```

```python
article.img.attrs['src']
```

```python
# url hence we have to make a get request to this url
```

```python
with open('inspiration.jpg', 'wb') as f:
    img_url = article.img.attrs['src']
    response = requests.get(img_url)
    
    f.write(response.content)
```

```python
print(len(article_element))
```

```python
for idx, article in enumerate(article_element):
    if(idx == 3):
        break
    with open('inspiration{}.jpg'.format(idx), 'wb') as f:
        img_url = article.img.attrs['src']
        response = requests.get(img_url)

        f.write(response.content)
```
