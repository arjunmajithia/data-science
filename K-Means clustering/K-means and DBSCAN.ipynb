```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples = 200, noise = 0.10)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X, y)
```

```python
centers = kmeans.cluster_centers_
pred = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c = pred)
plt.scatter(centers[:, 0], centers[:, 1], color = "red", marker = "*")
plt.title("K-Means fails for data which is not linearly separable")
plt.show()
```

```python
# DBSCAN PERFORMS WELL FOR SUCH DATA
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps = 0.20, min_samples = 5)
```

```python
pred = dbs.fit_predict(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = pred)
plt.show()
```
