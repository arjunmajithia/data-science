```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# DATASET PREPARATION
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, random_state = 3)
```

```python
print(X.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

```python
# K-MEANS IS AN UNSUPERVISED ALGORITHM HENCE WE WILL NOT USE y VALUES FOR TRAINING
```

```python
# RANDOM INITIALIZATION
k = 5
color = ["green", "red", "blue", "yellow", "orange"]

clusters = {}

for i in range(k):
    center = 10 * (2 * np.random.random((X.shape[1], )) - 1)
    
    points = []
    
    cluster = {
        'center': center,
        'points': points,
        'color': color[i]
    }
    
    clusters[i] = cluster
```

```python
# EUCLEDIAN DISTANCE BETWEEN CLUSTER CENTRE AND A POINT
def distance(d1, d2):
    return np.sqrt(np.sum((d1-d2)**2))
```

```python
# ASSIGN EVERY DATA POINT TO EACH CLUSTER
def assignPointToClusters(clusters):      # E-step (Expectation)
    for ix in range(X.shape[0]):
        dist = []
        curr_x = X[ix]

        for kx in range(k):
            d = distance(curr_x, clusters[kx]['center'])
            dist.append(d)

        current_cluster = np.argmin(dist)
        clusters[current_cluster]['points'].append(curr_x)
        
# UPDATE CENTER OF EVERY CLUSTER ACCORDING TO MEAN OF POINTS IN IT
def updateClusters(clusters):           # M-step (Maximization)
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        if pts.shape[0] > 0:
            new_u = pts.mean(axis = 0)
            clusters[kx]['center'] = new_u
            clusters[kx]['points'] = []
            
# PLOT THE CLUSTERS
def plotClusters(clusters):
    
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        # print(pts)
        
        try:
            plt.scatter(pts[:, 0], pts[:, 1], color = clusters[kx]['color'])
        except:
            pass
        
        uk = clusters[kx]['center']
        plt.scatter(uk[0], uk[1], color = "black", marker = "*")
        
    plt.show()
```

```python
assignPointToClusters(clusters)
plotClusters(clusters)
```

```python
# LOOP TO CONVERGE THE ALGORITHM

for i in range(50):
    assignPointToClusters(clusters)
    updateClusters(clusters)

assignPointToClusters(clusters)
plotClusters(clusters)
```
