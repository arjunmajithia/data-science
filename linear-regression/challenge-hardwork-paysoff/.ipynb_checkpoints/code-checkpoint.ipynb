```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv("X_Train.csv").values
Y = pd.read_csv("Y_Train.csv").values
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression(normalize = True)
lr.fit(X, Y)
```

```python
print(lr.predict([[4]])[0][0])
```

```python
plt.plot(X, lr.predict(X), c = 'orange')
plt.scatter(X, Y)
plt.show()
```

```python
from sklearn.externals import joblib
```

```python
joblib.dump(lr, 'hardwork_paysoff.pkl')
```
