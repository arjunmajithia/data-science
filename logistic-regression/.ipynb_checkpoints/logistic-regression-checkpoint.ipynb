```python
# preparing data

mean_01 = np.array([1,0.5])
cov_01 = np.array([[1,0.1],[0.1,1.2]])

mean_02 = np.array([4,5])
cov_02 = np.array([[1.21,0.1],[0.1,1.3]])


# Normal Distribution
dist_01 = np.random.multivariate_normal(mean_01,cov_01,500)
dist_02 = np.random.multivariate_normal(mean_02,cov_02,500)

print(dist_01.shape)
print(dist_02.shape)
```

```python
# creating training and test data
```

```python
data = np.zeros((1000, 3))
```

```python
data[:500,:2] = dist_01
data[500:,:2] = dist_02
data[500:,-1] = 1.0
```

```python
np.random.shuffle(data)
print(data[:10])
```

```python
split = int(0.8*data.shape[0])
```

```python
X_train = data[:split,:-1]
X_test = data[split:,:-1]

Y_train = data[:split,-1]
Y_test  = data[split:,-1]

print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

```python
# logistic regression
```

```python
# both of these calculations are for one example
def hypothesis(x, theta, bias):
    z = np.dot(x, theta) + bias
    return sigmoid(z)

def sigmoid(z):
    return 1.0/(1.0 + np.exp(-1.0 * z))

# calculating overall error
def error(X, y, theta, bias):
    m = X.shape[0]
    e = 0.0
    
    for i in range(m):
        y_ = hypothesis(X[i], theta, bias)
        e += y[i]*np.log2(y_) + (1-y[i])*np.log2(1-y_)
        
    return -e/m

# find out gradient for each feature
def gradient(X, y, theta, bias):
    grad_w = np.zeros(theta.shape)
    grad_b = 0.0
    
    m = X.shape[0]
    for i in range(m):
        y_ = hypothesis(X[i], theta, bias)
        
        grad_w += (y[i] - y_) * X[i]
        grad_b += (y[i] - y_)
        
    return [grad_w/m, grad_b/m]

def gradientDescent(X, y, learning_rate = 0.1, max_steps = 1000):
    m = X.shape[0]
    n = X.shape[1]
    
    theta_list = []
    error_list = []
    bias_list = []
    theta = 2*np.random.random((X_train.shape[1],))
    bias = 5*np.random.random()
    
    for i in range(max_steps):
        e = error(X, y, theta, bias)
        
        [grad_w, grad_b] = gradient(X, y, theta, bias)
        theta = theta + learning_rate * grad_w
        bias = bias + learning_rate * grad_b
        
        theta_list.append(theta)
        bias_list.append(bias)
        error_list.append(e)
        
    return theta_list, bias_list, error_list
        
```

```python
theta_list, bias_list, error_list = gradientDescent(X_train, Y_train)
```

```python
plt.plot(error_list)
plt.show()
print(theta_list[0])
```

```python
# prediction and accuracy
```

```python
def predict(x, theta, bias):
    y_ = hypothesis(x, theta, bias)
    
    if y_ < 0.5:
        return 0
    else:
        return 1
    
def accuracy(X, y, theta, bias):
    y_pred = []
    
    for i in range(y.shape[0]):
        p = predict(X[i], theta, bias)
        y_pred.append(p)
        
    y_pred = np.array(y_pred)
    
    return float((y_pred == y).sum())/y.shape[0]
```

```python
acc_list = []
for i in range(1000):
    a = accuracy(X_test, Y_test, theta_list[i], bias_list[i])
    acc_list.append(a)
```

```python
plt.plot(acc_list)
plt.show()
```
