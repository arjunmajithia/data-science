```python
import cv2
```

```python
# read an image
img = cv2.imread('dog.jpg')
# opencv reads image in form (B, G, R) instead of (R, G, B)
```

```python
import matplotlib.pyplot as plt
```

```python
# show image
plt.imshow(img)
plt.show()
```

```python
# hence we have to make it read as RGB
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(newImg)
plt.show()
```
