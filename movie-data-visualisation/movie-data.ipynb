```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('movie-metadata.csv')
```

```python
df.head(n = 10)
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
```

```python
#print(titles)
print(type(titles))
```

```python
d = {}                           # mapping length of movie title with number of movies

for title in titles:
    length = len(title)
    
    if d.get(length) is None:
        d[length] = 1
    else:
        d[length] += 1
```

```python
X = list(d.keys())
Y = list(d.values())
```

```python
plt.style.use('seaborn')

plt.scatter(X, Y)
plt.xlabel('Length of movie titles')
plt.ylabel('Number of movies')
plt.show()
```
