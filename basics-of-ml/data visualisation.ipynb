```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x, y1)                   # x-axis, y-axis
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Prices of fruits over time')
plt.show()
```

```python
plt.plot(x, y1, color = 'red', label = 'apple', linestyle = 'dashed')
plt.plot(x, y2, color = 'green', label = 'kiwi', marker = 'o')                   # x-axis, y-axis

# marker means from which points line is passing

plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Prices of fruits over time')

plt.legend()
plt.show()                                       # to show the output on the screen
```

```python
plt.plot(x, y1, color = 'green')
plt.show()

plt.plot(x, y2, color = 'red')                   # x-axis, y-axis
plt.show()                                       # to show the output on the screen
```

```python
# scatter plots
# when we do not a curve but only points
plt.scatter(x, y1)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
# adjust the size of the plot

plt.scatter(x, y1, color = 'red', label = 'apple', marker = '*')
plt.scatter(x, y2, color = 'green', label = 'kiwi', marker = '^')                   # x-axis, y-axis

# marker means from which points line is passing

plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Prices of fruits over time')

plt.legend()
plt.show()     
```

```python
# bar graphs
plt.bar([0, 1, 2], [10, 20, 15])        # first vector -> x coordinates, second vector -> height of bars
plt.show()
```

```python
# comapre bar graphs
x_coord = np.array([0, 1, 2])*2

plt.bar(x_coord - 0.25, [10, 20, 15], width = 0.5, label = "Current year", tick_label = ['Gold', 'Platinum', 'Silver'])     # first year prices
plt.bar(x_coord + 0.25, [20, 10, 12], width = 0.5, label = "Next year", color = 'red')     # second year

plt.title('Metals price comparison')
plt.xlabel('Metal')
plt.ylabel('Price')

plt.ylim(0, 40)                        # range of y axis

plt.legend()
plt.show()
```

```python
# normal distribution
```

```python
xsn = np.random.randn(100)           # gives 100 standard normally distributed numbers where sigma = 1, u = 0
# most values are in the range -1 to 1 
print(xsn)
```

```python
sigma = 5
u = 60

# to generate normally distributed numbers with mean = 60 and standard distribution as 5

X = np.round(xsn*sigma + u)
X2 = np.round(xsn*10 + 40)
print(X, X2)
```

```python
# histogram
```

```python
plt.hist(X, edgecolor='black', alpha = 0.7, label = 'Maths')
plt.hist(X2, edgecolor='black', alpha = 0.7, label = 'Physics')
# alpha denotes opacity of graph

plt.xlabel('Marks')
plt.ylabel('Frequency of Students')
plt.legend()
plt.show()
```

```python
# normal distribution 1-D
# in normal distribution more than 95% points lie in (u - 2*sigma, u + 2*sigma)
u = 5
sigma = 1

vals = u + sigma * np.random.randn(100)
print(vals.shape)

plt.hist(vals, 10)
plt.show()
```

```python
# multivaraite normal distribution
# it is normal distribution in multi dimensional space
# given by vector X = [x1, x2, ...... xn]
# mean u = [u1, u2, ....., un]
```

```python
mean = np.array([0.0, 0.0])
cov = np.array([[1, 0.8], [0.8, 1]])

mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1.3, 0.2], [0.2, 1.1]])

# generate multivariate normally distributed points having paramters as mean, covariance, number of points
dist = np.random.multivariate_normal(mean, cov, 500)
dist2 = np.random.multivariate_normal(mean2, cov2, 500)
print(dist.shape)

plt.scatter(dist[:, 0], dist[:, 1])
plt.scatter(dist2[:, 0], dist2[:, 1])
plt.show()
```
